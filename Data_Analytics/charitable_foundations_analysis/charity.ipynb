```python
category_mask = df['category'].replace({'Environment':0, 'Arts, Culture, Humanities':1, 'Religion':2,
       'Human Services':3, 'Education':4, 'Animals':5, 'International':6,
       'Health':7, 'Community Development':7, 'Human and Civil Rights':8,
       'Research and Public Policy':9})

category_mask = category_mask.to_frame()
category_mask
```

```python
# category_mask
score = df['score']
score_and_category = pd.concat([category_mask, score], axis=1)
```

```python
correlation_score_and_category = score_and_category.corr()
#таблицка с цветами корреляций
correlation_score_and_category.style.background_gradient(cmap='coolwarm')
```
